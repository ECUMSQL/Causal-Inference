# 基础复杂网络编程实践

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n= 10000
p= 0.001
ER=nx.erdos_renyi_graph(n,p) #创建一个ER随机网络
```

```python
# 获取平均度
d = dict(nx.degree(ER))
print("平均度：",sum(d.values())/len(ER.nodes()))
```

```python
# 获取所有可能的度值对应的概率
x = list(range(max(d.values())+1))
y = [i/n for i in nx.degree_histogram(ER)]  
```

```python
# 绘制度分布
plt.plot(x,y,'ro-')
plt.xlabel('$k$')
plt.ylabel('$P_k$')
```

# BA无标度网络为例子

```python
m =3
BA = nx.barabasi_albert_graph(n,m) #创建一个BA无标度网络
# 获取平均度
d = dict(nx.degree(BA))
print("平均度：",sum(d.values())/len(BA.nodes()))
```

```python
# 获取所有可能的度值对应的概率
x = list(range(max(d.values())+1))
y = [i/n for i in nx.degree_histogram(BA)] #这个其实就是找重复的数量和直方图类似
# 绘制度分布
plt.plot(x,y,'bo-')
plt.xlabel('$k$')
plt.ylabel('$P_k$')
```

```python
# 用双对数坐标系绘制度分布
plt.plot(x,y,'bo-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k$')
plt.ylabel('$P_k$')
```

```python
# 在双对数坐标轴下要把横坐标与纵坐标的0值排除
new_x = []
new_y = []
for i in range(len(x)):
    if y[i] != 0:
        new_x.append(x[i])
        new_y.append(y[i])
plt.plot(new_x,new_y,'bo-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k$')
plt.ylabel('$P_k$')
```
